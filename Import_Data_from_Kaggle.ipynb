#### Assuming we have imported 9000 zipped dataset from Kaggle, use the following process to unzip them all.
- Can also merge all data into a large dataset

```python
# Create a data dictionary
data_dict = {}

# Get the list of zip files
zip_files = !ls *.zip
zip_files = re.split(r'\s{2,}', zip_files[0])

start_time = time.time()
# Loop over each zip file
for filename in zip_files:
    print('------------------------------' + filename + '------------------------------')
    with zipfile.ZipFile(filename, 'r') as zip_ref:
        zip_ref.extractall()
        # Loop over each file within the zip file
        for file in zip_ref.namelist():
            if file.endswith('.csv'):
                key = os.path.splitext(file)[0]
                df = pd.read_csv(file)
                data_dict[key] = df
                print(df.head())

print("\n" + "--- %s seconds ---" % (time.time() - start_time))

# Check data dictionary)
print("\n")
print(data_dict.keys())
```

<br>

### Useful Functions

```python
# Remove non-English tokens - if needed
def english_only(text):
    words = re.findall(r'\b[a-zA-Z]+\b', text)
    return ' '.join(words)

# Apply to dataset
for data in data_dict.values():
    if 'page_text' in data.columns:
      data['page_text'] = data['page_text'].apply(english_only)

data_dict[list(data_dict.keys())[-1]]['page_text'] # Success
```
